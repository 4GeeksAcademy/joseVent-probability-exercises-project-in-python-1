### Exercise 1 

Two unbiased dice are thrown once and the total score is observed. Use a simulation to find the estimated probability that the total score is even or greater than 7

We could use a mathematical equation to solve this or we can run the experiment many times, and then count the number of times where the score is even or greater than 7

-Run the experiment 1000 times (roll 2 dice 1000 times, and sum the result)

-Keep track of the number of times that the sum was either greater than 7 or an even number

-Divide the number from step 2 by the number of iterations (1000)


```python
import random

# Initialize counters
count_even_or_greater_than_7 = 0
total_simulations = 1000

# Run the experiment 1000 times
for _ in range(total_simulations):
    # Roll two dice
    dice1 = random.randint(1, 6)
    dice2 = random.randint(1, 6)
    
    # Calculate the sum
    total_score = dice1 + dice2
    
    # Check if the sum is even or greater than 7
    if total_score % 2 == 0 or total_score > 7:
        count_even_or_greater_than_7 += 1

# Calculate the estimated probability
estimated_probability = count_even_or_greater_than_7 / total_simulations

# Print the result
print("Estimated probability:", estimated_probability)

```

### Exercise 2

A box contains 10 white balls, 20 reds and 30 greens. Draw 5 balls with replacement. What is the probability that:

a. 3 white or 2 red

b. All 5 are the same color


We will use the same system here. We will pick our balls during each round and count the number of times that:

a. 3 white or 2 red 

b. All 5 are the same color.

```python
#import libraries


import random
import numpy as np

# Initialize important variables
count_a = 0  # Counter for a. 3 white or 2 red
count_b = 0  # Counter for b. All 5 are the same color
total_simulations = 10000  # Number of simulations

# Run the experiment
for _ in range(total_simulations):
    # Make a list of the colors that we choose
    chosen_colors = []
    for _ in range(5):
        chosen_colors.append(random.choice(["white", "red", "green"]))
    
    # Convert it to a numpy array
    chosen_colors_array = np.array(chosen_colors)
    
    # Find the number of each color that we picked
    num_white = np.sum(chosen_colors_array == "white")
    num_red = np.sum(chosen_colors_array == "red")
    num_green = np.sum(chosen_colors_array == "green")
    
    # Keep track if the combination met the criteria
    if num_white >= 3 or num_red >= 2:
        count_a += 1
    
    if num_white == 5 or num_red == 5 or num_green == 5:
        count_b += 1

# Calculate probabilities
probability_a = count_a / total_simulations * 100
probability_b = count_b / total_simulations * 100

# Print results
print("The probability of 3 white or 2 red is:", probability_a, "%")
print("The probability of all the same color is:", probability_b, "%")

```
